```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('E:/New folder/train.csv')
data.head()
```

## Set index

```python
data.set_index(data['Loan_ID'], inplace=True)
data.drop('Loan_ID', axis=1, inplace=True)
```

```python
data.head(20)
```

## Remove NaN

```python
data['Credit_History'].fillna((data['Credit_History'].median()), inplace=True)
data['LoanAmount'].fillna((data['LoanAmount'].median()), inplace=True)
data['Loan_Amount_Term'].fillna((data['Loan_Amount_Term'].median()), inplace=True)
data['Gender'].fillna((data['Gender'].mode()[0]), inplace=True)
data['Married'].fillna((data['Married'].mode()[0]), inplace=True)
data['Dependents'].fillna((data['Dependents'].mode()[0]), inplace=True)
data['Self_Employed'].fillna((data['Self_Employed'].mode()[0]), inplace=True)
```

```python
data.head(30)
```

```python
data.isna().sum()
```

```python
data['Loan_Amount_Term'].value_counts()
```

## Convert categorical data to numerical

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
data['Dependents']=le.fit_transform(data['Dependents'])
```

```python
cat_data = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']
```

```python
for col in cat_data:
    data[col] = le.fit_transform(data[col])
```

```python
data.head(10)
```

## Split the data into training and testing data

```python
loan_x = data.iloc[:,1:-1]
loan_y = data.iloc[:,-1]
```

```python
loan_x
```

```python
loan_y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(loan_x, loan_y, train_size=0.8)
```

```python
print(X_train.shape)
print(X_test.shape)
```

## KNN Algorithm Visualization

```python
X_train.columns
```

```python
from mlxtend.plotting import plot_decision_regions
X_train=X_train[['Credit_History','Property_Area']]
print(X_train.shape)
print(y_train.shape)
```

```python
knn=KNeighborsClassifier()
knn.fit(X_train, y_train)
plot_decision_regions(X_train.to_numpy(), y_train.to_numpy(), clf=knn)
plt.xlabel('X_train')
plt.ylabel('y_train')
plt.title('KNN Algorithm')
plt.show()
```

## Bagging Classifier Visualization

```python
bag = BaggingClassifier()
bag.fit(X_train, y_train)
plot_decision_regions(X_train.to_numpy(), y_train.to_numpy(), clf=bag)
plt.xlabel('X_train')
plt.ylabel('y_train')
plt.title('Bagging Classification')
plt.show()
```

## Decision Tree Classifier

```python
X_train, X_test, y_train, y_test = train_test_split(loan_x, loan_y, train_size=0.8)
from sklearn.tree import DecisionTreeClassifier
```
